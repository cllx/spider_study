```python
import requests
from bs4 import BeautifulSoup
import re
import time
import selenium
from selenium import webdriver

job_ask_urllist = []
def get_page_url(url):
#     driver = webdriver.Firefox()
#     driver.get(url) #访问链接
#     time.sleep(3)
#     pagesource = driver.page_source #抓取网页源代码
#     driver.close()
    header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
    requests.encoding = 'utf-8'
    req = requests.get(url, headers=header)
    req.encoding = "utf-8"
    pagesource = req.text
    soup = BeautifulSoup(pagesource, 'lxml')
#     data1 = soup.select("div .pagenav")[0].find_all("a", class_="active")[0]
    page_urllist = []
    num = eval(soup.find_all("span", class_="lightblue total")[0].string)
    if num%10 == 0:
        page_num = num//10
    else:
        page_num = num//10 + 1
    for i in range(0, page_num):
        url = "https://hr.tencent.com/position.php?keywords=python&tid=0&start="+str(i*10)+"#a"
        page_urllist.append(url)
    return page_urllist
    
def get_job_ask_url(url):
    header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
    requests.encoding = 'utf-8'
    req = requests.get(url, headers=header)
    req.encoding = "utf-8"
    pagesource = req.text
    soup = BeautifulSoup(pagesource, 'lxml')
    data = soup.find_all("table", class_="tablelist")
    for line in data[0].find_all("tr", class_=["even", "odd"]):
        job_ask_urllist.append("https://hr.tencent.com/"+line.find_all("td")[0].a["href"])
        for data in line.find_all("td"):
            print(data.string)
    
def get_ask_info(url):
    header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
    requests.encoding = 'utf-8'
    req = requests.get(url, headers=header)
    req.encoding = "utf-8"
    pagesource = req.text
#     soup = BeautifulSoup(pagesource, 'lxml', from_encoding="utf-8")
    soup = BeautifulSoup(pagesource, 'lxml')
    data = soup.find_all("ul", class_="squareli")
    job_ask_info = ""
    for dataline in data:
        for linedata in dataline.find_all("li"):
            if linedata.string:
                job_ask_info = job_ask_info + linedata.string
            else:
                job_ask_info = job_ask_info + ""
    return job_ask_info

pageurl = get_page_url("https://hr.tencent.com/position.php?keywords=python&tid=0")
for url in pageurl:
    get_job_ask_url(url)
f = open("tencent_job.txt", "a+", encoding="utf-8")
for job_url in job_ask_urllist:
    job_info = get_ask_info(job_url)
    f.write(job_info.strip())
f.close()
```
