```python
# 生成中文验证码
from PIL import ImageDraw, ImageFont, ImageFilter, Image
import random
import sys
import math
import string
import codecs

font_path = "./day17/simkai.ttf"
# 位数
numbers = 4
# 验证码大小
size = (150, 60)
# 背景颜色
bgcolor = (255, 255, 255)
# 字体颜色
fontcolor = (0, 0, 0)
# 干扰线条
a = random.randint(0,255)
b = random.randint(0,255)
c = random.randint(0,255)
lingcolor = (a, b, c)
draw_line = True
line_numbers = (1, 5)

class RandomChar():
    @staticmethod
    def Unicode():
        val = random.randint(0x4E00, 0X9FBF) #汉字的unicode范围
        return unichr(val)
    
    @staticmethod
    def GB2312():
        head = random.randint(0xB0, 0xBF) #汉字的gb2312范围
        body = random.randint(0xA, 0xF)
        tail = random.randint(0, 0xF)
        val = (head << 8) | (body << 4) | tail
        gene_str = "%x" % val
        return codecs.decode(gene_str, 'hex').decode('gb2312', 'ignore') #这是python3的方式

def bgfontrange():
    global bgcolor
    color1 = random.randint(0,255)
    color2 = random.randint(0,255)
    color3 = random.randint(0,255)
    bgcolor = (color1,color2,color3)
    
def ftfontrange():
    global fontcolor
    color1 = random.randint(0,255)
    color2 = random.randint(0,255)
    color3 = random.randint(0,255)
    fontcolor = (color1,color2,color3)

def make_text():
    text = u""
    for i in range(4):
        text += RandomChar.GB2312()
    return text

# 随机画线
def make_line(draw, width, height):
    begin = (random.randint(0,width),random.randint(0,height))
    end = (random.randint(0,width),random.randint(0,height))
    draw.line([begin, end], fill=lingcolor, width=3) #绘线

# 生成验证码
def make_codepng():
    width, height = size #图片的宽度和高度
    image = Image.new("RGBA", (width, height), bgcolor) #创建图片
    draw = ImageDraw.Draw(image) #绘图工具
    text = make_text() #生成随机字符串
    font = ImageFont.truetype(font_path, 40) #字体
    font_width, font_height = font.getsize(text) #字体的宽高
    draw.text(((width-font_width)/numbers, (height-font_height)/2), 
              text, 
              font=font, 
              fill=fontcolor) #写入文字
#     画线
    if draw_line:
        num = random.randint(1,6)
        for i in range(num):
            make_line(draw, width,height)
    #(1,-0.1,0,-0.2,0.9,0)
    num1 = random.uniform(0.9, 1)
    num2 = random.uniform(-0.3, 0)
    num3 = random.uniform(-0.1, 0.1)
    num4 = random.uniform(-0.3, 0.1)
    num5 = random.uniform(0.8, 1)
    num6 = random.uniform(0, 0.2) #生成小数，min, max
    image = image.transform((width+30, height+20), 
                            Image.AFFINE, 
                            (num1,num2,num3,num4,num5,num6), 
                            Image.BILINEAR) #扭曲
    image = image.filter(ImageFilter.EDGE_ENHANCE_MORE)
    filename = r"./code/"+text+".png"
    with open(filename, "wb") as file:
        image.save(file, format="png")

for i in range(10):
    bgfontrange()
    ftfontrange()
    make_codepng()
```
