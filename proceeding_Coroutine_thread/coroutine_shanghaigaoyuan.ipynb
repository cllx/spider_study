```python
import requests
import selenium
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import gevent
import gevent.monkey
import urllib

gevent.monkey.patch_all() #自动切换、智能切换
def download(url, start, end, file):
    driver = selenium.webdriver.PhantomJS(r"./phantomjs.exe")
    driver.get(url)
    gevent.sleep(10)
    #0-100，100-200，200-300
    for i in range(start, end):
        js = "javascript:goPage('"+str(i)+"')" #实现js翻页
        driver.execute_script(js)
        print("js is run")
        gevent.sleep(10) #循环提取页面
        #提取页面数据
        soup = BeautifulSoup(driver.page_source, "lxml") #解析数据
        tabel = soup.find("table", attrs={"id": "report"})
        trs = tabel.find("tr").find_next_siblings()
        for tr in trs:
            tds = tr.find_all("td")
            linestr = "" #拼合数据
            for td in tds:
                linestr += td.text
                linestr += " # "
            linestr += "\r\n"
            print(linestr)
            file.write(linestr)
    driver.quit()
file = open("save.txt", "w", encoding="utf-8")
url = "http://www.hshfy.sh.cn/shfy/gweb2017/ktgg_search.jsp?zd=splc"
gevent.joinall([gevent.spawn(download, url, 0, 2, file),
                gevent.spawn(download, url, 2, 4, file),
                gevent.spawn(download, url, 4, 8, file),
                gevent.spawn(download, url, 8, 10, file),
                gevent.spawn(download, url, 10, 12, file)])
file.close()
```
