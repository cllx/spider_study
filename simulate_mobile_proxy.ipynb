```python
# encoding:utf-8
# header即是模拟手机浏览器的header
import urllib.request

def downloadAndroid(url):
    header = {"User-Agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Mobile Safari/537.36"}
    request = urllib.request.Request(url, headers=header)
    request.add_header("Connection", "keep-live")
    print(request.get_full_url()) #整个网页链接
#     print(request.get_host()) #服务器域名
    print(request.get_method()) #get或者post
#     print(request.get_type()) #http或者ftp,https
    response = urllib.request.urlopen(request)
    print(response.code) #200，404，403等
    print(response.info()) #网页详细信息
    print(response.read().decode('utf-8')) #网页内容
    
downloadAndroid("https://www.baidu.com/s?wd=python&rsv_spt=1&rsv_iqid=0xac504d1f000453be&issp=1&f=8&rsv_bp=0&rsv_idx=2&ie=utf-8&tn=baiduhome_pg&rsv_enter=1&rsv_sug3=7&rsv_sug1=2&rsv_sug7=100&rsv_sug2=0&inputT=1646&rsv_sug4=3476")
```

```python
# encoding: utf-8
# 这里的代理允许所有都可以访问，没有设置账号密码
import urllib.request
httpproxy = urllib.request.ProxyHandler({"http":"169.254.139.106:808"}) #代理无需账号
opener = urllib.request.build_opener(httpproxy) #创建一个打开器
request = urllib.request.Request("http://www.baidu.com/") #代理访问，url必须完整
response = opener.open(request) #打开网页，内置代理服务器
print(response.read().decode('utf-8'))
```
