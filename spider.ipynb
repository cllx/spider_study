```python
#图片爬取全代码
import requests
import os
url = "http://image.nationalgeographic.com.cn/2017/0211/20170211061910157.jpg"
root = "../spider_img/"
path = root + url.split('/')[-1]
try:
    if not os.path.exists(root):
        os.mkdir(root)
    if not os.path.exists(path):
        r = requests.get(url)
        with open(path, 'wb') as f:
            f.write(r.content)
            f.close()
            print("文件保存成功")
    else:
        print("文件已经存在")
except:
    print("爬取失败")
```

```python
#find_all的使用
import requests
import re
from bs4 import BeautifulSoup
r = requests.get("http://python123.io/ws/demo.html")
demo = r.text
soup = BeautifulSoup(demo,"html.parser")
print(soup.find_all('a'))
print("------------------------------------")
print(soup.find_all(['a','b']))
print("------------------------------------")
for tag in soup.find_all(True):
    print(tag.name)
    print("------------------------------------")
    #print(tag)
for tag in soup.find_all(re.compile('b')):
    print(tag.name)
print("------------------------------------")

#find_all检索标签属性值的字符串
#检索带有course属性值的p标签
print(soup.find_all('p', 'course'))
#查找id值为link1的标签及其内容
print("------------------------------------")
print(soup.find_all(id='link1'))
#查找id值为link的标签及其内容
print("------------------------------------")
print(soup.find_all(id='link'))
#查找id值为以link开头的标签及其内容
print("------------------------------------")
print(soup.find_all(id=re.compile('link')))
#recursive的使用(默认为Ture，检索对应内容下的所有子孙节点)
print("------------------------------------")
print(soup.find_all('a',recursive=False))
#string的使用，精确检索某个字符串
print("------------------------------------")
print(soup.find_all(string="Basic Python"))
#利用正则表达式检索Python的所有字符串
print("------------------------------------")
print(soup.find_all(string=re.compile('Python')))
```

```python
#中国大学排名定向爬虫
from bs4 import BeautifulSoup
import requests
import bs4

def getHTMLText(url):
    try:
        r = requests.get(url, timeout=30)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return "false"

def fillUnivList(ulist, html):
    soup = BeautifulSoup(html, "html.parser")
    for tr in soup.find('tbody').children:
        if isinstance(tr, bs4.element.Tag):
            tds = tr('td')
            ulist.append([tds[0].string, tds[1].string, tds[2].string, tds[3].string])

def printUnivList(ulist, num):
    #tplt = "{0:^10}\t{1:^10}\t{2:^10}\t{0:^10}"
    print("{:^10}\t{:^6}\t{:^10}\t{:^10}".format("排名", "学校名称", "地区", "总分"))
    #print(tplt.format("排名", "学校名称", "地区", "总分", chr(12288)))
    for i in range(num):
        u = ulist[i]
        print("{:^10}\t{:^6}\t{:^10}\t{:^10}".format(u[0], u[1], u[2], u[3]))
        #print(tplt.format(u[0], u[1], u[2], u[3], chr(12288)))
    
def main():
    uinfo = []
    url = "http://www.zuihaodaxue.cn/zuihaodaxuepaiming2018.html"
    html = getHTMLText(url)
    fillUnivList(uinfo, html)
    printUnivList(uinfo, 20) #列出20所学校的相关信息

main()
```
