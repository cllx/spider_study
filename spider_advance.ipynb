```python
# import urllib2 #python3中将urllib和urllib2合并了，用下面的方式使用
import urllib.request
def download1(url):
    return urllib.request.urlopen(url).read() #读取全部网页

def download2(url):
    return urllib.request.urlopen(url).readlines() #读取每一行网页数据，压入到列表

def download3(url):
    response = urllib.request.urlopen(url) #网页抽象为文件
    while True:
        line = response.readline() #读取一行
        if not line:
            break
        print(line)

url = "http://www.baidu.com" #urlopen只能处理http，不能处理https
print(download1(url))
print(download2(url))
print(download3(url))
```

```python
# 这属于直接调用浏览器进行爬取
# import urllib2 #python3中将urllib和urllib2合并了，用下面的方式使用
# 网页屏蔽 服务器屏蔽

import selenium #网页测试框架
# import selenium.webdriver #模拟浏览器
from selenium import webdriver
import re

def getnumberbyname(searchname):
    # pass
    url = "https://search.51job.com/list/030200,000000,0000,00,9,99,"+searchname+",2,1.html?lang=c&stype=&postchannel=0000&workyear=99&cotype=99&degreefrom=99&jobterm=99&companysize=99&providesalary=99&lonlat=0%2C0&radius=-1&ord_field=0&confirmdate=9&fromType=&dibiaoid=0&address=&line=&specialarea=00&from=&welfare="
    # driver = webdriver.Firefox(executable_path = 'E:\firefox\Mozilla Firefox\firefox.exe') #调用火狐浏览器
    driver = webdriver.Firefox()
    driver.get(url) #访问链接
    pagesource = driver.page_source #抓取网页源代码
#     print(pagesource)
#     restr = '<span class="download__text">(.*?)</span>' #带括号只取括号里面的内容
#     restr = '<div class="rt">共(\\d+)条职位</div>' #直接这样抓(爬)不到，是因为对应位置可能有空白和换行符
    restr = '<div class="rt">([\S\s]*?)</div>'
    regex = re.compile(restr, re.IGNORECASE)
    mylist = regex.findall(pagesource)
    driver.close() #关闭
    newstr = mylist[0].strip()
    restr2 = '共(\\d+)条职位'
    regex2 = re.compile(restr2, re.IGNORECASE)
    mylist2 = regex2.findall(newstr)
    if len(mylist2) == 0:
        print("失败")
    else:
        return mylist2[0]

pythonlist = ["python", "python 运维", "python 测试", "python 数据", "python web"]
for pystr in pythonlist:
    print(pystr, getnumberbyname(pystr))
```

```python
# 这属于直接调用浏览器进行爬取
# uth-8 urllib2 == urllib.request
# import urllib2
import urllib.error
import urllib.request

def download(url):
    response = urllib.request.urlopen(url, timeout=10) #请求超时10s
    print(type(response))
    print(response.info) #打印整个网站请求的完整的信息
    print(response.read()) #read()读取全部网页，read(100)读取网页的前100个内容
    print(response.read(100)) #read()读取全部网页，read(100)读取网页的前100个内容

# 有时候不是程序的问题，可能是网络或网站本身的问题，所以在爬网站的时候通常要使用try except来显示异常，不影响爬虫程序的使用
try:
    print(download("http://www.baidu.com"))
except urllib.error.HTTPError as e:
    print(e.code)
except urllib.error.URLError as e:
    print(e.reason)
```

```python
# 这属于伪装成浏览器，也可以伪装成手机浏览器，chrom浏览器也有手机模式
import urllib.request
def download(url):
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
    request = urllib.request.Request(url, headers=headers)
    request.add_header("Connection", "keep-alive")
#     data = urllib.request.urlopen(request).read()
    response = urllib.request.urlopen(request)
    data = response.read().decode("utf-8")
    print(response.code)
#     return urllib.request.urlopen(url).read() #读取全部网页
    return data

url = "https://sou.zhaopin.com/?jl=489&sf=100001&st=999999&kw=Java&kt=3"
print(download(url))
```

```python
# 这属于伪装成浏览器，也可以伪装成手机浏览器，chrom浏览器也有手机模式
import urllib.request
from urllib.parse import quote,unquote
def download(addr, mytype):
    url = "http://sou.zhaopin.com/jobs/searchresult.ashx?"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
    word = {"j1": addr, "kw": mytype}
    word = urllib.parse.urlencode(word)
    url = url + word
    request = urllib.request.Request(url, headers=headers)
    request.add_header("Connection", "keep-alive")
#     data = urllib.request.urlopen(request).read()
    response = urllib.request.urlopen(request)
    data = response.read().decode("utf-8")
    print(response.code)
#     return urllib.request.urlopen(url).read() #读取全部网页
    return data

addr = "深圳"
mytype = "python"
print(download(addr, mytype))
```

```python
# encoding:utf-8
# post方式爬取数据
import urllib.request
# data = "first=true&p=1&kd=python"
data = urllib.parse.urlencode({'first': "true", 'p': 1, 'kd': "python"}).encode('utf-8')
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
requests = urllib.request.Request("https://www.lagou.com/jobs/positionAjax.json", headers=headers)
requests.data = data #post操作
print(urllib.request.urlopen(requests).read().decode("utf-8"))
```
