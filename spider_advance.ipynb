```python
# import urllib2 #python3urlliburllib2
import urllib.request
def download1(url):
    return urllib.request.urlopen(url).read() #

def download2(url):
    return urllib.request.urlopen(url).readlines() #

def download3(url):
    response = urllib.request.urlopen(url) #
    while True:
        line = response.readline() #
        if not line:
            break
        print(line)

url = "http://www.baidu.com" #urlopenhttphttps
print(download1(url))
print(download2(url))
print(download3(url))
```

```python
# 
# uth-8 urllib2 == urllib.request
# import urllib2
import urllib.error
import urllib.request

def download(url):
    response = urllib.request.urlopen(url, timeout=10) #10s
    print(type(response))
    print(response.info) #
    print(response.read()) #read()read(100)100
    print(response.read(100)) #read()read(100)100

# try except
try:
    print(download("http://www.baidu.com"))
except urllib.error.HTTPError as e:
    print(e.code)
except urllib.error.URLError as e:
    print(e.reason)
```

```python
# chrom
import urllib.request
from urllib.parse import quote,unquote
def download(addr, mytype):
    url = "http://sou.zhaopin.com/jobs/searchresult.ashx?"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
    word = {"j1": addr, "kw": mytype}
    word = urllib.parse.urlencode(word)
    url = url + word
    request = urllib.request.Request(url, headers=headers)
    request.add_header("Connection", "keep-alive")
#     data = urllib.request.urlopen(request).read()
    response = urllib.request.urlopen(request)
    data = response.read().decode("utf-8")
    print(response.code)
#     return urllib.request.urlopen(url).read() #
    return data

addr = ""
mytype = "python"
print(download(addr, mytype))
```
